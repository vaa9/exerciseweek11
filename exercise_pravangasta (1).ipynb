```python
df["member_date"] = df["Member_number"].astype(str) + df["Date"].astype(str)
df["member_date"] 
```

<h2> Data Preprocessing

Ubahlah dataset di atas menjadi bentuk one hot encoding. Asumsikanlah bahwa pembelian barang yang dilakukan oleh Customer ID yang sama dalam hari yang sama terjadi dalam 1 transaksi yang sama (satu Order ID)



```python
df_group = df['member_date']+df['itemDescription']
df_group = pd.DataFrame(df_group,columns=['member_date','itemDescription'])
df_group
```

```python
df_group = df.groupby(['member_date', 'itemDescription'])['itemDescription'].count().unstack().fillna(0)
df_group
```

```python
def change_value(encode):
  if encode >= 1 :
    return 1
  else: 
    return 0

df_group = df_group.applymap(change_value)
df_group
```

## Apriori Algorithm

Gunakanlah algroitma apriori untuk mencari tahu produk mana yang sering dibeli secara bersama-sama (min support = 0.01)

```python
from mlxtend.frequent_patterns import apriori, association_rules
freq_items = apriori(df_group,use_colnames=True, min_support=0.01)
freq_items.head(7)
```

Berdasarkan hasil di atas, buatlah association rules nya dengan syarat min confidence = 0.1 

```python
rules = association_rules(freq_items, metric="confidence", min_threshold=0.1)
rules.head()
```
